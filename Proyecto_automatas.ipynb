# **EJECUTAR CÓDIGO A PSEUDOCÓDIGO**


````python
import re

def python_a_pseudocodigo(codigo_python):
    # Expresiones regulares para patrones específicos en el código Python
    expresiones = [
        (r'\bdef\b +((?![0-9])\w+) *\((.*?)\) *:', r'Procedimiento \1(\2)'),
        (r'\bif\b +(.*):', r'Si \1 entonces'),
        (r'\belif\b +(.*?):', r'SiNo \1 entonces'),
        (r'(\belse\b):', r'SiNo'),
        (r'\bfor\b\s+(\w+)\s+\bin\s+(.*?):', r'Para \1 en \2 hacer'),
        (r'\bwhile\b\s+(.*?):', r'Mientras \1 hacer'),
        (r'\breturn\b(.*)', r'Devolver \1'),
        (r'\bprint\b(.*)', r'Escribir \1'),
        (r'(\b(?![0-9])\w+\b) *= *[^=](.+)', r'leer \1 = \2')
    ]

    pseudocodigo = codigo_python

    # Traduce el código línea por línea
    for expresion, reemplazar in expresiones:
        pseudocodigo = re.sub(expresion, reemplazar, pseudocodigo, flags=re.MULTILINE)

    return pseudocodigo

# Ejemplo de código Python
codigo_python = '''
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(5))

j = True
while j:
  j = False
  print("while")

for i in range(3):
  print("for")
  if i == 2:
    print("if")
  elif i==1:
    print("1")
  else:
    print("else")
'''

pseudocodigo = python_a_pseudocodigo(codigo_python)
# Salida
print(pseudocodigo)
````

````python
import re

def pseudocodigo_a_python(pseudocodigo):
    # Expresiones regulares para patrones específicos en el código Python
    expresiones = [
        (r'\bProcedimiento\b +((?![0-9])\w+) *\((.*?)\)', r'def \1(\2):'),
        (r'\bSi\b +(.*) +\bentonces\b', r'if \1:'),
        (r'\bSiNo\b +(.*?) +\bentonces\b', r'elif \1:'),
        (r'\bSiNo\b', r'else:'),
        (r'\bPara\b +(\w+) +en +(.+?) +hacer', r'for \1 in \2:'),
        (r'\bMientras\b +(.+?) +hacer', r'while \1:'),
        (r'\bDevolver\b +(.*)', r'return \1'),
        (r'\bEscribir\b *(.*)', r'print\1'),
        (r'\bleer\b +(\b(?![0-9])\w+\b) *= *(.+)', r'\1 = \2')

        # Solo Dios y yo entendemos como funciona esto, y a mi se me esta olvidando
    ]

    codigo_python = pseudocodigo

    # Traduce el código línea por línea
    for expresion, reemplazar in expresiones:
        codigo_python = re.sub(expresion, reemplazar, codigo_python, flags=re.MULTILINE)

    return codigo_python

# Ejemplo de código Python
pseudocodigo = '''
Procedimiento factorial(n)
  Si n == 0 entonces
    Devolver  1
  SiNo
    Devolver  n * factorial(n-1)

Escribir (factorial(5))

leer j = True
Mientras j hacer
  leer j = False
  Escribir ("while")

Para i en range(3) hacer
  Escribir ("for")
  Si i == 2 entonces
    Escribir ("if")
  SiNo i==1 entonces
    Escribir ("1")
  SiNo
    Escribir ("else")
'''

codigo_python = pseudocodigo_a_python(pseudocodigo)
# Salida
print(codigo_python)
````

````python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(5))

j = True
while j:
  j = False
  print("while")

for i in range(3):
  print("for")
  if i == 2:
    print("if")
  elif i==1:
    print("1")
  else:
    print("else")
````
